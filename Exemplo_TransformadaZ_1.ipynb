```python
# Relação entre os polos de F(z) e a evolução temporal de f(k)
a = 0.5;
k = np.linspace(0, 10, 11)
fk = a**k;
print("fk = \n",fk)

plt.subplot(1,2,1)
plt.plot(k, fk,'*')
plt.title("Evolução Temporal de f(k)")
plt.xlabel("k")
plt.ylabel("f(k)")
plt.grid(True)

v = [-1.1,1.1,-1.1,1.1];
x = np.linspace(-1,1,100);
y = (1-x**2)**.5;             # x^2+y^2 = 1

plt.subplot(1,2,2)
plt.plot(x,y,'b',x,-y,'b',a,0,'*')
plt.title("Plano Z")
plt.axis(v);
plt.grid(True)

plt.show()
```

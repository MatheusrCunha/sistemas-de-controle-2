```python
import numpy as np
import control
import math
import cmath
import matplotlib.pyplot as plt
```

```python
V3 = 10;
Vo = 100
R = 23;
C = 1000e-6;
NT = 23;

Io = 16;            # valor inicial
T = 5e-3;             # período de amostragem
RC = R*C;    # constante RC
Tf = NT*10e-3;                 # tempo final de simulação
Kmax = int(Tf/T+1);
a = np.exp(-T/RC);
b = 1 - np.exp(-T/RC);
print("b = \n", b)
print("a = \n",a)
```

```python
#t = np.linspace(0, 2, 999);
t = np.arange(0, Tf + Tf/999, Tf/999)

V1 = Vo*np.exp(-t/RC);
plt.figure()

plt.plot(t, V1)
plt.title("V1(t)")
plt.grid(True)
plt.show()
```

```python
k = np.linspace(0, Tf/T, Kmax);
print("k = \n",k)
V2 = Vo*np.exp(-k*T/RC);
#print("V2 = \n", V2)
plt.figure()
plt.title("Resposta do sistema")
plt.grid(True)
plt.plot(t,V1,k*T, V2,'*')
plt.show()
```

```python
I3 = np.zeros(Kmax);
I3[0] = 100;   # para k = 0

for j in range(1,Kmax):
    I3[j]=a*I3[j-1];

print("I3 = \n",I3)
plt.figure()
plt.title("Resposta do sistema")
plt.grid(True)
plt.plot(t,V1,k*T, I3,'o')
plt.show()
```

```python
t = np.arange(0, Tf + Tf/999, Tf/999)

V4 = V3*(1 - np.exp(-t/RC));
plt.figure()

plt.plot(t, V4)
plt.title("V4(t)")
plt.grid(True)
plt.show()
```

```python
k = np.linspace(0, Tf/T, Kmax);
print("k = \n",k)
V5 = V3*(1 - np.exp(-k*T/RC));
#print("V2 = \n", V2)
plt.figure()
plt.title("Resposta do sistema")
plt.grid(True)
plt.plot(k*T, V5,'*')
plt.show()
```

```python
Vc4 = np.zeros(Kmax);
Vc4[0] = 0;   # para k = 0

print("K = \n", Kmax)

for j in range(1,Kmax):
     Vc4[j]=a*Vc4[j-1] + b*V3;

#print("Vc4 = \n",Vc4)
plt.figure()
plt.title("Resposta do sistema")
plt.grid(True)
plt.plot(k*T, Vc4,'o')
plt.show()
```

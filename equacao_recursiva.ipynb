```python
import numpy as np
import control
import math
import cmath
import matplotlib.pyplot as plt
```

```python
Io = 16;            # valor inicial
T = .1;             # período de amostragem
RC = 0.14426951;    # constante RC
Tf = 2;             # tempo final de simulação
Kmax = int(Tf/T+1);
a = np.exp(-T/RC);
print("a = \n",a)
```

```python
t = np.linspace(0, 2, 999);
I1 = Io*np.exp(-t/RC);
plt.figure()

plt.plot(t, I1)
plt.show()
```

```python
k = np.linspace(0, Tf/T, Kmax);
print("k = \n",k)
I2 = Io*np.exp(-k*T/RC);
print("I2 = \n",I2)
plt.figure()
plt.title("Resposta do sistema")
plt.grid(True)
plt.plot(t,I1,k*T, I2,'*')
plt.show()
```

```python
I3 = np.zeros(Kmax);
I3[0] = Io;   # para k = 0

for j in range(1,Kmax):
    I3[j]=a*I3[j-1];

print("I3 = \n",I3)
plt.figure()
plt.title("Resposta do sistema")
plt.grid(True)
plt.plot(t,I1,k*T, I3,'o')
plt.show()
```
